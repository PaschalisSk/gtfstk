```python
from __future__ import division
import datetime as dt
from collections import OrderedDict
import sys, os
import dateutil.relativedelta as rd
from pathlib import Path

import pandas as pd
import numpy as np


DIR = Path('..')
sys.path.append(str(DIR))
import gtfstk as gt

DATA_DIR = DIR/'data/'

%load_ext autoreload
%autoreload 2
# Comment out the next line if matplotlib is not installed
%matplotlib inline  
```

```python
# Import a feed
feed = gt.read_gtfs(DATA_DIR/'cairns_gtfs.zip', dist_units='km')

# Set study date
date = gt.compute_busiest_date(feed, gt.get_first_week(feed))
date
```

```python
gt.assess(feed)
```

```python
# Compute trip stats
trip_stats = gt.compute_trip_stats(feed)
trip_stats.head().T
```

```python
# Add shape distance traveled field to stop times
feed = gt.append_dist_to_stop_times(feed, trip_stats)
feed.stop_times.head().T
```

```python
# Get trips
gt.get_trips(feed, date).head()
```

```python
# Compute stop stats for that date
stop_stats = gt.compute_stop_stats(feed, date, split_directions=True)
stop_stats.head()
```

```python
# Compute stop time series for that date
sts = gt.compute_stop_time_series(feed, date, split_directions=True)
sts = gt.downsample(sts, freq='6H')
sts.head().T
```

```python
# Compute route stats for that date
route_stats = gt.compute_route_stats(feed, trip_stats, date, split_directions=True)
route_stats.head().T
```

```python
gt.plot_headways(route_stats)
```

```python
# Compute route time series
rts = gt.compute_route_time_series(feed, trip_stats, date, split_directions=True)
rts = gt.downsample(rts, freq='6H')
rts.head()
```

```python
fts = gt.compute_feed_time_series(feed, trip_stats, date, freq='1H')
gt.downsample(fts, freq='4H').head()
```

```python
gt.plot_feed_time_series(fts)
```

```python
# Get trip locations for every hour of the given date
rng = pd.date_range('1/1/2000', periods=24, freq='H')
times = [t.strftime('%H:%M:%S') for t in rng]
loc = gt.compute_trip_locations(feed, date, times)
loc.head()
```

```python
# Get a route timetable
route = feed.routes['route_id'].iat[0]
gt.get_route_timetable(feed, route, date).head().T
```

```python
# Get a stop time table
stop = feed.stops['stop_id'].iat[0]
gt.get_stop_timetable(feed, stop, date).head().T
```
