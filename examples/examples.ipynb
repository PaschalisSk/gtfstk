```python
from __future__ import division
import datetime as dt
from collections import OrderedDict
import sys, os
import dateutil.relativedelta as rd

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

gtfstk_dir = '../../gtfstk/'
sys.path.append(gtfstk_dir)
import gtfstk as gt

%load_ext autoreload
%autoreload 2
%matplotlib inline  
```

```python
# Import a feed
feed = gt.Feed(gtfstk_dir + 'data/cairns_gtfs.zip')

# Set study date
date = feed.get_busiest_date(feed.get_first_week())
date
```

```python
# Get trips stats
trips_stats = feed.compute_trips_stats()
trips_stats
```

```python
# Add shape distance traveled field to stop times
feed.add_dist_to_stop_times(trips_stats)
feed.stop_times
```

```python
# Get trips
feed.get_trips(date)
```

```python
# Get stops stats for that date
stops_stats = feed.get_stops_stats(date, split_directions=True)
stops_stats
```

```python
sts = feed.compute_stops_time_series(date, split_directions=True)
sts = gt.downsample(sts, freq='6H')
sts
```

```python
# Get routes stats for that date
routes_stats = feed.compute_routes_stats(trips_stats, date, split_directions=True)
routes_stats
```

```python
# Plot route headways
gt.plot_headways(routes_stats, max_headway_limit=None)
```

```python
# Get routes time series
rts = feed.compute_routes_time_series(trips_stats, date, split_directions=True)
rts = gt.downsample(rts, freq='6H')
rts
```

```python
f = feed.compute_feed_time_series(trips_stats, date, freq='1H')
gt.downsample(f, freq='4H')
```

```python
# Get trip locations for every hour of the given date
rng = pd.date_range('1/1/2000', periods=24, freq='H')
times = [t.strftime('%H:%M:%S') for t in rng]
loc = feed.compute_trips_locations(date, times)
loc
```

```python
# Get a route timetable
route = feed.routes['route_id'].iat[0]
feed.get_route_timetable(route, date)
```

```python
# Get a stop time table
stop = feed.stops['stop_id'].iat[0]
feed.get_stop_timetable(stop, date)
```
