```python
from __future__ import division
import datetime as dt
from collections import OrderedDict
import sys, os
import dateutil.relativedelta as rd

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%load_ext autoreload
%autoreload 2
%matplotlib inline  

# Set your gtfs_toolkit path here
gtfs_toolkit_path = '../../gtfs-toolkit/'
sys.path.append(gtfs_toolkit_path)
from gtfs_toolkit.feed import *
```

```python
# Import a feed
feed_path = gtfs_toolkit_path + 'data/cairns_gtfs.zip'
feed = Feed(feed_path)
```

```python
# Get trips stats
trips_stats = feed.get_trips_stats()
trips_stats
```

```python
# Add shape distance traveled field to stop times
feed.add_dist_to_stop_times(trips_stats)
```

```python
# Get trips activity for the busiest day of the first week of the feed
date = feed.get_busiest_date_of_first_week()
print(date)
feed.get_trips(date)
```

```python
# Get stops stats for that date
stops_stats = feed.get_stops_stats(date, split_directions=True)
stops_stats
```

```python
sts = feed.get_stops_time_series(date, split_directions=True)
sts = downsample(sts, freq='6H')
sts
```

```python
# Get routes stats for that date
routes_stats = feed.get_routes_stats(trips_stats, date, split_directions=True)
routes_stats
```

```python
# Plot route headways
plot_headways(routes_stats, max_headway_limit=None)
```

```python
# Get routes time series
rts = feed.get_routes_time_series(trips_stats, date, split_directions=True)
rts = downsample(rts, freq='6H')
rts
```

```python
arts = agg_routes_time_series(rts)
arts
```

```python
# Get vehicles locations for every hour of the given date
linestring_by_shape = feed.get_linestring_by_shape()
print(date)
rng = pd.date_range('1/1/2000', periods=24, freq='H')
times = [t.strftime('%H:%M:%S') for t in rng]
loc = feed.get_vehicles_locations(linestring_by_shape, date, times)
loc
```

```python
# Create and dump all stats to CSV
feed.dump_all_stats(gtfs_toolkit_path + 'examples/stats/', split_directions=True, date=date)
```
