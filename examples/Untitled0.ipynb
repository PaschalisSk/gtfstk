```python
import datetime as dt
import pandas as pd
import numpy as np
```

```python
rng = pd.date_range('1/1/2011', periods=24, freq='H')
mi = [(dt,i, j) for dt in rng for i in range(5) for j in range(2)]
ts = pd.DataFrame(np.random.randn(len(mi), 2), index=pd.MultiIndex.from_tuples(mi, names=['time', 'route', 'direction']), columns=['A', 'B'])
ts
```

```python
ts.unstack(['route', 'direction']).resample('12H', how=sum).stack()
```
