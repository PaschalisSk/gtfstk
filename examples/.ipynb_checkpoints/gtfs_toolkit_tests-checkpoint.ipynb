```python
# Get routes time series for first workweek
dates = feed.get_first_week()
#trips_stats = feed.get_trips_stats()
trips_stats
# rs = feed.get_routes_stats(trips_stats, dates)
# rs
rts = feed.get_routes_time_series(trips_stats, dates, freq='5Min')
rts
```

```python
#pd.options.display.mpl_style = 'default'
f = utils.downsample(rts, '1H')
f
utils.plot_routes_time_series(f)
```

```python
frames = list(rts.values())
def process_index(k):
    return tuple(k.rsplit('-', 1))
new_frames = []
for f in frames:
    ft = f.T
    ft.index = pd.MultiIndex.from_tuples([process_index(k) for k,v in ft.iterrows()], 
      names=['route_id', 'direction_id'])
    new_frames.append(ft.T)
g = pd.concat(new_frames, axis=1, keys=list(rts.keys()))
g.columns
g['mean_daily_duration'].resample('6H', how=np.sum)#, how={'mean_daily_duration': np.sum})
```
