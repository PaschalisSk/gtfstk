```python
# Plot route headways
f0 = routes_stats[routes_stats['direction_id'] == 0].\
  set_index('route_id').sort_index(ascending=False)
f1 = routes_stats[routes_stats['direction_id'] == 1].\
  set_index('route_id').sort_index(ascending=False)
data_frames = [f0, f1]

colors = ['red', 'blue'] 
alpha = 0.7
columns = ['max_headway', 'mean_headway']
titles = ['Max Headway','Mean Headway']
ylabels = ['stop_id', 'stop_id']
xlabels = ['minutes', 'minutes']
fig, axes = plt.subplots(nrows=1, ncols=2)
for (i, column) in enumerate(columns):
    F = pd.DataFrame({'direction_0': f0[column]/60, 'direction_1': f1[column]/60})
    F.plot(kind='barh', ax=axes[i], color=colors, alpha=alpha, figsize=(10, 10))
    axes[i].set_title(titles[i])
    axes[i].set_xlabel(xlabels[i])
    axes[i].set_ylabel(ylabels[i])
fig.tight_layout() 
```

```python
#pd.options.display.mpl_style = 'default'
f = utils.downsample(rts, '1H')
f
utils.plot_time_series(f)
```

```python
frames = list(rts.values())
def process_index(k):
    return tuple(k.rsplit('-', 1))
new_frames = []
for f in frames:
    ft = f.T
    ft.index = pd.MultiIndex.from_tuples([process_index(k) for k,v in ft.iterrows()], 
      names=['route_id', 'direction_id'])
    new_frames.append(ft.T)
g = pd.concat(new_frames, axis=1, keys=list(rts.keys()))
g.columns
g['mean_daily_duration'].resample('6H', how=np.sum)#, how={'mean_daily_duration': np.sum})
```
