```python
def my9(group):
    if group.shape[0] > 9:
    # Sort group ascending by start time
        group = group.sort(['start_date', 'start_time'])
        group['9free'].iloc[9:] = 0
    return group

i = 0
for index, group in f.groupby('smartcard_id'):
    if group.shape[0] > 9 and i < 3:
        i += 1
        print(group['9free'])

print('='*40)
i = 0
g = f.groupby('smartcard_id', group_keys=False).apply(my9)
for index, group in g.groupby('smartcard_id'):
    if group.shape[0] > 9 and i < 3:
        i += 1
        print(group['9free'])
print('='*40)
for index, group in g.groupby('smartcard_id'):
    if group.shape[0] > 9:
        if sum(group['9free']) > 9:
            print('Fail')
print('OK')
print(g.shape)
```
