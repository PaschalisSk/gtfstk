```python
# Load Auckland feed
feed = Feed('/Users/araichev/Desktop/auckland_20131029_no_school_routes.zip')
```

```python
# Get trips stats
trips_stats = feed.get_trips_stats()
trips_stats
```

```python
f = pd.merge(feed.routes, trips_stats)[['route_id', 'route_short_name', 'direction_id', 'trip_id', 'num_stops', 'duration', 'distance']]
f
g = f.groupby(['route_short_name', 'direction_id']).mean().reset_index()
g.columns = ['route_short_name', 'direction_id', 'num_stops_per_trip', 'duration_per_trip (h)', 'distance_per_trip (km)']
g
g.to_csv('/Users/araichev/Desktop/auckland_routes_stats.csv', index=False)
```
